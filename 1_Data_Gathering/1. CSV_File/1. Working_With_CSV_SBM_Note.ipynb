## 14. Convert column value to abbreviation

```python
# you can convert a string "Royal Challengers Bangalore" to RCB
# first create a function 
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    if name == "Kolkata Knight Riders":
        return "KKR"
    else:
        return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename,'team2':rename})
```

### 24.2 New column creation

```python
df_w = pd.read_csv("weatherdata.csv", header =0)
#print(df_w.head())

# MaxTemp is in Centegrade , I want a new column "Maxtemp_F" in F ( data["MaxTemp"] * 9/5 +32 )
df_w["Maxtemp_F"] = df_w["MaxTemp"] * 9/5 +32 
df_w["Mintemp_F"] = df_w["MinTemp"] * 9/5 +32 
df_w.head()
```

#### <font color='red'>New Column Creation From Formula by Function and map()? </font>

```python
def c_2_f_converter(temp_c):
    return temp_c * 9/5 +32
```

```python
df_w = pd.read_csv("weatherdata.csv", header =0)
df_w["Maxtemp_F_F"]= df_w["MaxTemp"].map(c_2_f_converter) # Uning user defined function c_2_f_converter
df_w
```
