## Stacked bar chart

### Example 1

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from https://allisonhorst.github.io/palmerpenguins/
species = ('Adelie', 'Chinstrap', 'Gentoo')
weight_counts = {
    "Below": np.array([70, 31, 58]), # Number of penguins below average body mass.
    "Above": np.array([82, 37, 66]), #Number of penguins above average body mass.
}
width = 0.5
bottom = np.zeros(3)  # Initialize bottom positions for stacking

# Create bar chart
for boolean, weight_count in weight_counts.items():
    #print(weight_counts.items())
    plt.bar(species, weight_count, width, label=boolean, bottom=bottom) #1st tile it will plot bar for Below Key of Dictionary [70, 31, 58]
    bottom += weight_count  # Update bottom for stacking # here it will increase bottom array to [70, 31, 58] ; bottom[0,0,0] + [70, 31, 58]

# Title and legend
plt.title("Number of penguins with above average body mass")
plt.legend(loc="upper right")

# Show plot
plt.show()
```

### Example 2

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
species = ('Adelie', 'Chinstrap', 'Gentoo')
sex_counts = {
    'Male': np.array([73, 34, 61]),
    'Female': np.array([73, 34, 58]),
}
width = 0.6  # Bar width

bottom = np.zeros(3)  # Initialize bottom positions for stacking

# Create bar chart
for sex, sex_count in sex_counts.items():
    p = plt.bar(species, sex_count, width, label=sex, bottom=bottom)
    bottom += sex_count  # Update bottom for stacking

    # Add bar labels
    plt.bar_label(p, label_type='center')

# Title and legend
plt.title('Number of penguins by sex')
plt.legend()
```

##### Bar Chart - Plot sales across each product category

1. Adding labels to Axes
2. Reducing the bar width
3. Giving Title to the chart
4. Modifying the ticks to show information in (million dollars)

```python
import numpy as np
import matplotlib.pyplot as plt

product_category = np.array(['Furniture', 'Technology', 'Office Supplies'])
sales = np.array ([4110451.90, 4744557.50, 3787492.52] )

# plotting bar chart and setting bar width to 0.5 and aligning it to center
plt.bar(product_category, sales, width= 0.5, align='center', edgecolor='Orange',color='cyan')

# Adding and formatting title
plt.title("Sales Across Product Categories\n", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})

# Labeling Axes
plt.xlabel("Product Category", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
plt.ylabel("Sales", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})

# Modifying the ticks to show information in (million dollars)
ticks = np.arange(0, 6000000, 1000000)
labels = ["{}M".format(i//1000000) for i in ticks]
plt.yticks(ticks, labels)

plt.show()
```

### A Box and Whisker Plot (or Box Plot) is a convenient way of visually displaying the data distribution through their quartiles. The lines extending parallel from the boxes are known as the whiskers, which are used to indicate variability outside the upper and lower quartiles. Outliers are sometimes plotted as individual dots that are in-line with whiskers. Box Plots can be drawn either vertically or horizontally.

### Box plots divide the data range into three important categories, which are as follows:

### Median value: This is the value that divides the data range into two equal halves, i.e., the 50th percentile.Interquartile range (IQR): These data points range between the 25th and 75th percentile values. Outliers: These are data points that differ significantly from other observations and lie beyond the whiskers.
    
### Here are the types of observations one can make from viewing a Box Plot:

### 1. What the key values are, such as: the average, median, 25th percentile, etc.
### 2. If there are any outliers and what their values are.
### 3. If the data is symmetrical or not.
### 4. How tightly is the data grouped.
### 5. If the data is skewed and if so, in what direction.
### Two of the most commonly used variation of Box Plot are: variable-width Box Plots and notched Box Plots

### Find Q1, Q3 and IQR

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
list_1 = [48.49, 67.54, 57.47, 68.17, 51.18, 68.31, 50.33, 66.7, 
          45.62, 43.59, 53.64, 70.08, 47.69, 61.27, 44.14, 51.62, 
          48.72, 65.11]

# Compute Q1 and Q3
Q1 = np.percentile(list_1, 25)  # 25th percentile
Q3 = np.percentile(list_1, 75)  # 75th percentile

# Compute IQR
IQR = Q3 - Q1

# Print the results
print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")

# Create a box plot
plt.boxplot(list_1, vert=False, patch_artist=True)
plt.title("Box Plot of Given Data")
plt.xlabel("Values")
plt.show()
```

# SubPlots - Plot Sales for various markets for years 2012 to 2015

```python
import numpy as np
import matplotlib.pyplot as plt

years = np.array(['2012', '2013', '2014', '2015'])

sales_africa = np.array([127187.27, 144480.70, 229068.79, 283036.44])

sales_USCA = np.array([492756.60, 486629.30, 627634.98, 757108.13])

sales_LATAM = np.array([385098.15, 464733.29, 608140.77, 706632.93])

sales_Asia_Pacific = np.array([713658.22, 863983.97, 1092231.65, 1372784.40])

sales_Europe = np.array([540750.63, 717611.40, 848670.24, 1180303.95])
```

##### Subplots are shown in the same graph as line charts, identified by different colours

```python
fig, ax = plt.subplots() # it returns a blank figure and axex (x,y)
```

```python
fig, ax = plt.subplots() # it returns a blank figure and axex (x,y)
# now inside the blank figure we want a line chart to be printed, keeping axes as blank 
europe, = ax.plot(years, sales_Europe) # years = x axis, sales = y axis

europe.set_label("Europe") # set the label name for this plot


# 2nd line graph in same figure 
usca, = ax.plot(years, sales_USCA) # years = x axis, sales = y axis
usca.set_label("USCA") # set the label name for this plot

#Print the label legends
plt.legend()
```

```python
fig, ax = plt.subplots()

europe, = ax.plot(years, sales_Europe)
europe.set_label("Europe")

usca, = ax.plot(years, sales_USCA)
usca.set_label("USCA")
usca.set_dashes([2, 2, 2, 2]) # 2 point line, 2 point brteak, 2 point line, 2 point brteak

africa, = ax.plot(years, sales_africa)
africa.set_label("Africa")
africa.set_dashes([2, 2, 5, 2]) # 2 point line, 2 point brteak, 5 point line, 2 point brteak

asia, = ax.plot(years, sales_Asia_Pacific)
asia.set_label("Asia Pacific")
asia.set_dashes([2, 2, 10, 2])  # 2 point line, 2 point brteak, 10 point line, 2 point brteak

latam, = ax.plot(years, sales_LATAM)
latam.set_label("LATAM")
latam.set_dashes([2, 5, 5, 2])  # 2 point line, 5 point brteak, 5 point line, 2 point brteak

plt.legend()

plt.legend(bbox_to_anchor=(1.31,0.4)) # to bring the legend box out of the main figure box, it is anchoring with main box

plt.show()
```

# Subplots are shown in seperate graphs. Each chart can be of different types.

```python
# dividing the subplot into 3 columns and 2 rows and all division will have seperate X and Y axis having same values
# 1st row  ax[0][0], ax[0][1], ax[0][2]
# 2nd row  ax[1][0], ax[1][1], ax[1][2]
fig, ax = plt.subplots(ncols=3, nrows=2, sharex=True, sharey=True)
# plt.subplot(4,4,2) # means it is a subplot of 1st row and column 2 position in  figure with 4 rows and 4 columns
```
